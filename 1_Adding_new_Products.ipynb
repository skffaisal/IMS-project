## Read Data from JSON

```python
import json
fd = open("record.json",'r')
r = fd.read()
fd.close()

record = json.loads(r)
```

```python
record # this is the list of  minimum 30 products with 5 attributes
```

## Add New Item into Inventory

```python
a=int(input(" press 1 if you want to upgrade an existing invetoty \n press 2 if you want add a new item in inventory \n"))
```

```python
if (a==1):
  prod_id = str(input("Enter product id:"))
  name = str(input("Enter name:"))
  pr = int(input("Enter price:"))
  a = record[prod_id]['qn']                                                                       # Upgrading existing inventory
  qn = int(input("Enter quantity:"))
  category=str(input("Enter the category :"))
  Weight=str(input("Enter the Weight:"))
  record[prod_id] = {'name': name, 'pr': pr, 'qn': qn+a,'category':category ,'Weight':Weight}
  a=0


if (a==2):
  prod_id = str(input("Enter product id:"))
  name = str(input("Enter name:"))
  pr = int(input("Enter price:"))                                                                # Adding new items in inventory
  qn = int(input("Enter quantity:"))
  category=str(input("Enter the category :"))
  Weight=str(input("Enter the Weight:"))

  record[prod_id] = {'name': name, 'pr': pr, 'qn': qn,'category':category ,'Weight':Weight}
```

```python
js = json.dumps(record)               # Updating the record file

fd = open("record.json",'w')
fd.write(js)
fd.close()
```
